## Logistic Regression | Full Case-study (Facebook Ads)


### Importing Dataset & Extracting Features


```python
fb_ads = pd.read_csv('data/social-network-ads.csv')
fb_ads
```

```python
purchased = fb_ads[fb_ads['Purchased'] == 1]
purchased
```

```python
not_purchased = fb_ads[fb_ads['Purchased'] == 0]
not_purchased
```

### Data Exploration


```python
sns.scatterplot(data=fb_ads, x=fb_ads['Age'], y=fb_ads['EstimatedSalary'], hue=fb_ads['Purchased'])
```

### Data Splitting & Preprocessing


```python
X = fb_ads.drop('Purchased', axis=1).values
y = fb_ads['Purchased'].values
```

```python
from sklearn.preprocessing import StandardScaler
sc = StandardScaler()
sc.fit_transform(X)
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=0)
```

### Model Training & Predicting


```python
from sklearn.linear_model import LogisticRegression
```

```python
clf = LogisticRegression()
```

```python
clf.fit(X_train, y_train)
```

### Model Evaluation & Prediction 


```python
y_pred = clf.predict(X_test)
```

```python
from sklearn.metrics import confusion_matrix, accuracy_score
```

```python
print(accuracy_score(y_test, y_pred))
print(confusion_matrix(y_test, y_pred))
```
